```python
#Observation de la population par ville
#info sur l'âge et la catégorie socio-professionnelle par ville

import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
#import data

dataVilleInsee=pd.read_csv('/Users/ysewanono/Desktop/HackatonPython/BTX_TD_POP6_2011.csv', sep=';')
#dataVilleInsee=dataVilleInsee.dropna()
dataVilleInsee[:5]
```

```python
df=dataVilleInsee[dataVilleInsee.columns[2:len(dataVilleInsee.columns)]]
#print df[:0]
df['POP TOTALE']= df.sum(axis=1)
dataVilleInsee['POP TOTALE']=df['POP TOTALE']
```

```python
categories=[]
for column in dataVilleInsee[:0]:
    column=column.replace('CS1_8','').replace('AGEQ65','').replace('SEXE','').replace('*POP2011','')
    newcolumns=column.split("_")
    #newcolumns=newcolumns[2:]
    #newcolumns=newcolumns.pop()
    categories.append(newcolumns[0])


#print dataVilleInsee.columns   
#print categories
dataVilleInsee.columns= categories
#print dataVilleInsee[:0]
print dataVilleInsee[:4]


df2=dataVilleInsee[['3']]
df2['SUM CADRES']=df2.sum(axis=1)
dataVilleInsee['SUM CADRES']=df2['SUM CADRES']

dataVilleInsee['POURCENTAGE CADRE']=(df2['SUM CADRES']/df['POP TOTALE'])*100
```

```python
#SORT DATA
villeparpop=dataVilleInsee[['CODGEO','LIBGEO','POP TOTALE','SUM CADRES','POURCENTAGE CADRE']]
villeparpop=villeparpop[villeparpop['POP TOTALE']>100000]
villetriepourcadre= villeparpop.sort(['POURCENTAGE CADRE','POP TOTALE'],ascending=[False,False])
print villetriepourcadre[:10]
```

```python
#POURCENTAGE DE CHOMEURS
df3=dataVilleInsee[['8']]
df3['SUM CHOMEURS']=df3.sum(axis=1)
dataVilleInsee['SUM CHOMEURS']=df3['SUM CHOMEURS']

dataVilleInsee['POURCENTAGE CHOMEURS']=(df3['SUM CHOMEURS']/df['POP TOTALE'])*100
```

```python
villeparchomeurs=dataVilleInsee[['CODGEO','LIBGEO','POP TOTALE','SUM CHOMEURS','POURCENTAGE CHOMEURS']]
villeparchomeurs=villeparchomeurs[villeparchomeurs['POP TOTALE']>100000]
villeparchomeurs=villeparchomeurs.sort(['POURCENTAGE CHOMEURS','POP TOTALE'],ascending=[False,False])
print villeparchomeurs[:10]
```

```python
#POURCENTAGE DE RETRAITES #ON A DIT "CADRES DYNAMIQUES"
df4=dataVilleInsee[['7']]
df4['SUM RETRAITE']=df4.sum(axis=1)
dataVilleInsee['SUM RETRAITE']=df4['SUM RETRAITE']

dataVilleInsee['POURCENTAGE RETRAITE']=(df4['SUM RETRAITE']/df['POP TOTALE'])*100
```
